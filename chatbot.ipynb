```python
import re

# Intent dictionary
INTENTS = {
    "greeting": ["hi", "hello", "hey"],
    "goodbye": ["bye", "goodbye"],
    "capabilities": ["capabilities", "what can you do"],
    "hours": ["hours", "open"],
    "help": ["help", "support"]
}

RESPONSES = {
    "greeting": "Hello! I am SimpleBot. Type 'capabilities' to see what I can do.",
    "goodbye": "Goodbye! Have a great day.",
    "capabilities": "I can respond to greetings, tell my capabilities, provide working hours, and help instructions.",
    "hours": "Our support hours are Monday-Friday, 9am-5pm EST.",
    "help": "Try asking: 'What can you do?' or 'What are your hours?'",
    "fallback": "Sorry, I did not understand that. Please try again or type 'help'."
}

def detect_intent(user_input):
    user_input = user_input.lower()
    for intent, keywords in INTENTS.items():
        for word in keywords:
            if word in user_input:
                return intent
    return "fallback"

def chatbot():
    print("SimpleBot is running. Type 'exit' to stop.")
    while True:
        user_input = input("You: ")
        if user_input.lower() == "exit":
            print("Bot:", RESPONSES["goodbye"])
            break
        
        intent = detect_intent(user_input)
        print("Bot:", RESPONSES[intent])

chatbot()
```
